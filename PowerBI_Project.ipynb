## 3. Data Report

Data Preparation, Cleaning, Analysis and Visualization can all be executed using multiple different platforms. This project demonstrated that both Tableau and Microsoft Power BI can produce similar figures. Intuitively, both platforms seem to offer different **advantages and disadvantages**.

Tableau is great for live interactive dashboards, analyzing larger datasets. However, using Tableau has a higher computational cost, as well as weaker native data modelling. On the other hand, Power BI has Excel/Azure integration, which is beneficial for, e.g. using the Power Query Editor. However, Power BI is less ideal for huge or non-Microsoft environments and is Windows-focused.

The choice of which platform is more suitable is dependent on the task, company, as well as given resources for fulfilling the given task.
